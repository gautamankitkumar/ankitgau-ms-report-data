# Goal

To simulate surface excess values for a ternary alloy of CuAgAu for all compositions using analytical equations

## Introduction

Surface excess for a multicomponent alloy can be predicted using Langmuir-McLean equations. This equations can be written as follows:

\begin{align}
\frac{x_{1}^{s}}{1- x_{1}^{s} - x_{2}^{s}} &=  \frac{x_{1}^{b}}{1- x_{1}^{b} - x_{2}^{b}} exp\Big(- \frac{\Delta G_{seg}^{(1)}}{RT}\Big)\\
\frac{x_{2}^{s}}{1- x_{1}^{s} - x_{2}^{s}} &=  \frac{x_{2}^{b}}{1- x_{1}^{b} - x_{2}^{b}} exp\Big(- \frac{\Delta G_{seg}^{(2)}}{RT}\Big)
\end{align}

where, 
- $x_{i}^{s}$  in the equation denotes surface composition (mole fraction) of component $i$
- $x_{i}^{b}$ represents its bulk composition -
- $\Delta G_{seg}^{(i)}$ in all of these equations represents free energy change related to segregation of component $i$ to the surface for one mole of atoms. 
- $T$ is the temperature in Kelvin.

These equations are modeled in [zhao-2020](https://doi.org/10.1002/admi.201901784). The authors have used the equations to model Pd‐Cu‐Ag and Pd‐Cu‐Mo ternary system. We demonstrate the use of the equations for CuAgAu system with parameters as mentioned in [zhao-2018](https://doi.org/10.1016/j.ijhydene.2017.12.039)

Full equations can be found the linked articles. The parameters are enumerated here for reference.

| Element 	&nbsp; &nbsp; | V ($\frac{cm^3}{mol}$) 	&nbsp; &nbsp;| $\phi ^*$ / V 	&nbsp; &nbsp;| $n_{ws}^{1/3}$ &nbsp; &nbsp;	| $\gamma$ ($\frac{mJ}{m^2}$) &nbsp; &nbsp;	| $K$ (GPa) 	&nbsp; &nbsp;| $G$ (GPa) 	&nbsp; &nbsp;	|
|:-:	|:-:	|:-:	|:-:	|:-:	|:-:	|:-:	|
| Pd 	| 8.89 	| 5.45 	| 1.67 	| 2050 	| 180 	| 44 	|
| Ag 	| 10.25 	| 4.35 	| 1.36 	| 1250 	| 100 	| 30 	|
| Au 	| 10.19 	| 5.15 	| 1.57 	| 1500 	| 220 	| 27 	|
| Cu 	| 7.12 	| 4.45 	| 1.47 	| 1825 	| 140 	| 48 	|
| Ni 	| 6.60 	| 5.20 	| 1.75 	| 2450 	| 180 	| 76 	|
| Pt 	| 9.10 	| 5.65 	| 1.78 	| 2475 	| 230 	| 61 	|
| H 	| 1.69 	| 5.20 	| 1.50 	| / 	| / 	| / 	|

where,
- V is atomic volume
- $\phi ^*$ is chemical potential for electronic charge
- $n_{ws}$ is the electron density at the boundary
- $\gamma$ is surface energy of pure metal
- $K$ is bulk modulus of pure metal
- $G$ is shear modulus of pure metal

## Methods

The two equations represent a set of non-linear equations where $\Delta G$ itself is a function of $x_{i}^{b}$s. The equations are written start with an initial guess of $x_{i}^{s}$s and re-iterated until we get both sides of the equations to be equal.

The equations are programmed via `scipy.optimize.fsolve`

```python
! pip install mpltern
import ternary
import numpy as np
import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.optimize import fsolve, minimize
```

```python
def get_surface_excess(x_bulk):
    """
    Takes three bulk composition and returns surface composition using McLean equations
    
    x_bulk: tuple of three floats representing bulk values
    
    """
    # Incoming order is Cu, Ag, Au
    # Order within this code should be x1b = Ag, x2b = Au, x3b = Cu , Thus rearranged
    
    x3b, x1b, x2b = x_bulk

    T = 600  # Kelvin

    # Bulk moduli of solute atoms, i.e. Ag, Au
    K = np.array([100, 220])  # GPa

    # Shear moduli of solvent, Cu
    G = 48  # GPa

    # Molar volume cm3/mol
    V = np.array([10.25, 10.19, 7.12])

    # Molar Surface area
    
    A = np.pi ** (1 / 3) * (6 * V) ** (2 / 3)

    # rad = (3 * (V * 1e-6 / N_A) / 4 / np.pi) ** (1 / 3)  # radius in m
    # A = 4 * np.pi * rad ** 2 * N_A  # rad in m2/mol
    # print(V, rad, A)
    # lattice constant in m
    # a = np.array([0.409, 0.408, 0.365])*1e-9
    # r = a/2/np.sqrt(2) # radii in m

    
    # Surface energies in mJ/m2
    surf_en = np.array([1250, 1500, 1825])

    # Enthalpy of mixing in J/mol
    # H12 = -4648
    # H23 = -1704
    # H13 = 757

    # Electron Density at the boundary
    eta = np.array([1.36, 1.57, 1.47])

    # Chemical Potential for electronic charge
    phi = np.array([4.35, 5.15, 4.45])

    # For FCC111
    z_l = 6  # In-layer Coordination
    z_v = 3  # Half of Out-of-plane Coordination
    z = z_l + 2 * z_v  # Total coordination in Bulk

    # Defining solute strain energy. E[0] for Ag in Cu. E[1] for Au in Cu
    E = np.array([(2 * K[i] * G * (V[i] - V[2]) ** 2) / (3 * K[i] * V[2] + 4 * G * V[i]) for i in range(2)])

    P = 12.35
    Q = 9.4 * P
    alpha = 0.04

    def getH_mix(X, ind1, ind2):
        """
        Gets H mixing of ind1 in ind2 with current surface composition in X
        """
        xa = X[ind1]
        xb = X[ind2]
        na = eta[ind1]
        nb = eta[ind2]
        ca = xa / (xa + xb)
        cb = xb / (xa + xb)
        pv = phi * V

        def get_values(y):
            cas, fab, Va = y

            V_new = V[ind1] ** (2 / 3) * (1 + alpha * fab * (pv[ind1] - pv[ind2]))  # Returns Va**(2/3)

            cs_new = ca * Va / (ca * Va + cb * V[ind2] ** (2 / 3))

            f_new = (1 - cas) * (1 + 8 * (cas ** 2) * (1 - cas) ** 2)

            return [cas - cs_new, fab - f_new, Va - V_new]

        guess = np.array([ca, ((1 - ca) * (1 + 8 * (ca ** 2) * (1 - ca) ** 2)), V[ind1] ** (2 / 3)])
        
        
        sol = fsolve(get_values, guess)

        ca_final, f_ab, V_a = sol  # Returns  Va**(2/3)

        nw_diff = na - nb
        nw_avg = (na + nb) / 2
        del_H = V_a * (-P * (pv[ind1] - pv[ind2]) ** 2 + Q * nw_diff ** 2) / nw_avg

        # Returns H_sol, f*H_int
        return [del_H, f_ab * del_H]


    i1, i2 = 0, 1
    w12 = 1 / z * (getH_mix([x1b, x2b, x3b], i1, i2)[1] + getH_mix([x1b, x2b, x3b], i2, i1)[1])

    i1, i2 = 0, 2
    w13 = 1 / z * (getH_mix([x1b, x2b, x3b], i1, i2)[1] + getH_mix([x1b, x2b, x3b], i2, i1)[1])

    i1, i2 = 1, 2
    w23 = 1 / z * (getH_mix([x1b, x2b, x3b], i1, i2)[1] + getH_mix([x1b, x2b, x3b], i2, i1)[1])

    # Hmix in J/mol
    #wHmix23 = (-28821 - 2468 * (x2b / (x2b + x3b)) + 9541 * (x2b / (x2b + x3b)) ** 2)
    
    # Some approximation for enthalpy of mixing can be obtained from 
    # 1. https://link.springer.com/content/pdf/10.1007/BF02657857.pdf
    # 2. https://link.springer.com/content/pdf/10.1007/BF02644402.pdf 
    # print(wHmix23 / z, w23)

    wprime = w12 - w13 - w23


    def getSurfaceConcentration(x):
        # Surface concentration of Ag, Au
        x1s, x2s = x

        term1 = surf_en[0] * A[0] - surf_en[2] * A[2]
        term2 = 2 * w13 * (z_l * (x1b - x1s) + z_v * (x1b - 0.5))
        term3 = wprime * (x2s - x2b - z_v * x2b)
        term4 = E[0]
        H1seg = term1 + term2 + term3 - term4
        # print(term1, term2, term3, term4)

        term1_ = surf_en[1] * A[1] - surf_en[2] * A[2]
        term2_ = 2 * w23 * (z_l * (x2b - x2s) + z_v * (x2b - 0.5))
        term3_ = wprime * (x1s - x1b - z_v * x1b)
        term4_ = E[1]
        H2seg = term1_ + term2_ + term3_ - term4_
        # print(term1_,term2_,term3_,term4_)
        eqn1 = x1s / (1 - x1s - x2s) - x1b / (1 - x1b - x2b) * np.exp(-H1seg / 8.314 / T)
        eqn2 = x2s / (1 - x1s - x2s) - x2b / (1 - x1b - x2b) * np.exp(-H2seg / 8.314 / T)

        return [eqn1, eqn2]


    res = fsolve(getSurfaceConcentration, np.array([x1b, x2b]))
    # Return excess 
                      # Cu,               # Ag     # Au      # Initial composition in order Cu, Ag, Au
    return (np.array([1- res[0] - res[1], res[0], res[1]]) - np.array(x_bulk)) * 100
```

```python
def ag_surf(p):
    # Gets second element, denoting Ag excess
    return get_surface_excess(p)[1]

def au_surf(p):
    # Gets third element, denoting Au excess
    return get_surface_excess(p)[2]

def cu_surf(p):
    # First element, denoting Cu excess
    return get_surface_excess(p)[0]

v_min = 100
v_max = -100
for i in np.linspace(0,1,11):
    for j in np.linspace(0,1-i,11):
        k = 1.0-i-j
        v_min = min(v_min,*get_surface_excess([i,j,1-i-j]))
        v_max = max(v_max,*get_surface_excess([i,j,1-i-j]))
v_min,v_max
```

```python
get_surface_excess([0.92,0.01,0.07]) # Just a test
```

```python
# Define a dense grid of points to plot on a ternary diagram

a, b = np.mgrid[0:1:100j, 0:1:100j]
mask = a + b <= 1
a, b = a[mask], b[mask]
a,b

cu_, ag_, au_ = [], [], []
for (x,y) in zip(a,b):
    cu_.append(cu_surf([x,y,1-x-y]))
    ag_.append(ag_surf([x,y,1-x-y]))
    au_.append(au_surf([x,y,1-x-y]))

data = [cu_,ag_,au_]
```

```python
import mpltern

%matplotlib inline

cm = 'bwr'

# Combine all functions
data_array = [cu_surf,ag_surf,au_surf]


fig = plt.figure(figsize=(19,10))
fig.subplots_adjust(wspace=0.5)
norm1 = mpl.colors.Normalize(vmin=v_min, vmax = v_max)
sm1 = plt.cm.ScalarMappable(cmap=cm, norm = norm1)

# Define an empty list to collect all the axes object. Used for Colorbar later
axes = []
level_list = [
    [-5],
    [10],
    [-10]
]

fs = 20 # fontsize
for i in range(3):
    ax = fig.add_subplot(1,3,i+1,projection='ternary')
    axes.append(ax)
    ax.tricontourf(a,b,(1-a-b),data[i],cmap=cm,norm = norm1,levels=np.linspace(v_min,v_max),antialiased=True)
    cx = ax.tricontour(a, b, (1-a-b), data[i],levels=level_list[i],colors='k',alpha=0.8,origin='upper')
    clabels = plt.clabel(cx,fontsize=fs,use_clabeltext=True,fmt='%1.0f')
    ax.set_tlabel("Cu", fontsize=fs)
    ax.set_llabel("Ag", fontsize=fs)
    ax.set_rlabel("Au", fontsize=fs)
    ax.tick_params(labelsize=fs-5,labelrotation='horizontal')
    ax.taxis.set_label_rotation_mode('horizontal')
    ax.laxis.set_label_rotation_mode('horizontal')
    ax.raxis.set_label_rotation_mode('horizontal')
    ax.set_title(['Cu Excess','Ag Excess', 'Au Excess'][i],fontsize=fs+3,pad = 80)
    for l in clabels:
        l.set_rotation(0)
    
fig.suptitle('Analytical Surface Excess Plots',y=0.87,fontsize=25)
cb = fig.colorbar(sm1, ax = axes, orientation="horizontal")
cb.ax.tick_params(labelsize=fs)

#https://github.com/pydata/xarray/issues/3275 Colorbar size
cb.set_label(label='Surface Excess (Final % - Bulk %)', size=20)
# plt.savefig('analytical.pdf',dpi=300)
plt.show()
```

# Conclusion

The analytical solution qualitatively captures the segregation trend of Ag to the surface. However, the quantitative values do not match both of the Monte Carlo simulations and earlier published values. Zhao et. al have included a parameter fitted to an experimentally observed values for the case of PdCuAg.  

More work is needed in to better predict the  segregation energy of the alloys.
